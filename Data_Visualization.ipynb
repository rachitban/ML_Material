**Import**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

**Iris Dataset Analysis**

```python
df = sns.load_dataset('iris')
df
```

**Histograms** 

Histograms represent the distribution of a continuous variable over a given interval or period of time. Histograms plot the data by dividing it into intervals called ‘bins’. 

```python
sns.histplot(x=df.sepal_length)
```

```python
sns.histplot(x='sepal_length',data=df,bins=12)
```

**KDE**

```python
sns.kdeplot(x=df.sepal_length)
```

**Scatter Plot**

Scatter plots are used to determine the relationship between two variables. They show how much one variable is affected by another. It is the most commonly used data visualization technique and helps in drawing useful insights when comparing two variables. The relationship between two variables is called correlation. Two variables must be numeric.




```python
sns.scatterplot(x=df.petal_width,y=df.petal_length)
```

**Joint Plot**


Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sns.jointplot(x=df.petal_width,y=df.petal_length,kind='reg')
```

**Pair Plot**

It is used to plot multiple pairwise bivariate (two variable) distribution in a dataset. It creates a matrix and plots the relationship for each pair of columns. It also draws a univariate distribution for each variable on the diagonal axes.

```python
sns.pairplot(data=df)
```

**Count Plot**

This is a seaborn-specific function which is used to plot the count or frequency distribution of each unique observation in the categorical variable. It is similar to a histogram over a categorical rather than quantitative variable.



```python
sns.countplot(x=df.species)
```

**Box Plot**

Box plots, also known as box and whisker plots are used to visualize variations and compare different categories in a given set of data.In a box and whisker plot:

1. the box spans the interquartile range   
2. a vertical line inside the box represents the median.
3. two lines outside the box, the whiskers extends either to max and min observation or it could be any complex strategy like some % of IQR.





```python
sns.boxplot(y=df.petal_length, x=df.species)
```

**t-sne**

Used to visualize higher dimension data.

```python
from sklearn.manifold import TSNE

m = TSNE(n_components=2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

df1=df.drop('species',axis=1)

scaled_data = StandardScaler().fit_transform(df1)

tsne_features = m.fit_transform(df1)
```

```python
x = tsne_features[:,0]
y = tsne_features[:,1]

sns.scatterplot(x=x, y=y,hue=df.species)
```
